```python
import numpy as np 
import pandas as pd 

import os
for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        print(os.path.join(dirname, filename))
```

**Import Library**

```python
from sklearn import preprocessing
from sklearn.model_selection import train_test_split
from keras.models import Sequential
from keras.layers import Input, Dense
```

**Load Dataset**

```python
dataset_path = "/kaggle/input/oranges-vs-grapefruit/citrus.csv"
df = pd.read_csv(dataset_path)
df.head()
```

```python
df.info()
```

```python
# Mengubah kolom kategorikal (tipe data onject) menjadi numerik
df.loc[df["name"] == "grapefruit", "name"] = 0
df.loc[df["name"] == "orange", "name"] = 1
```

```python
# Mengubah kolom kategorikal (tipe data onject) menjadi numerik dengan LabelEncoder
label_encoder = preprocessing.LabelEncoder()
df['name'] = label_encoder.fit_transform(df['name'])
print(df)
```

```python
df.info()
```

```python
 # Mengubah dataset yang masih dalam bentuk dataframe menjadi array
dataset = df.values
print(dataset)
```

```python
# Memisahkan atribut dependen (label/y) dan atribut independen (fitur/x) 
X = dataset[:,1:6]

'''pilih 5 kolom terakhir sebagai atribut 
bilangan sebelum koma untuk memilih baris pada dataframe 
bilangan setelah koma untuk memilih kolom pada dataframe'''
```

```python
# Memisahkan label dari dataset (label di baris pertama)
y = dataset[:,0]
```

**Normalisasi**

```python
# Mengubah skala menjadi 0-1
min_max_scaler = preprocessing.MinMaxScaler()
X_scale = min_max_scaler.fit_transform(X)
X_scale
```

**Split Data**

```python
# Memisahkah data training dan testing 70:30
X_train, X_test, Y_train, Y_test = train_test_split(X_scale, y, test_size=0.3)
```

**Modelling**

```python
model = Sequential([
    Input(shape=(5,)), 
    Dense(32, activation='relu'),
    Dense(32, activation='relu'),
    Dense(1, activation='sigmoid')
])
```

```python
model.compile(optimizer='sgd',
             loss='binary_crossentropy',
             metrics=['accuracy'])
```

**Training**

```python
model.fit(X_train, Y_train, epochs=100)
```

**Testing**

```python
model.evaluate(X_test, Y_test, batch_size=1)
```

Model dapat melalukan prodiksi dengan sangat baik (akurasi 92%) pada data yang belum pernah dilihat
